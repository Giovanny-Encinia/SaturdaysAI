**3. What is the percentage of German citizens (*native-country* feature)?**

```python
# You code here
German = data[data['native-country'] == 'Germany']
GermanNumber = German["age"].count()
PercGerman = GermanNumber / data["age"].count()
print("% of German citizens: ", PercGerman * 100, "%")
```

**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year? **

```python
# You code here
Less50K = data[data['salary'] == '<=50K']
More50K = data[data['salary'] == '>50K']
MeanLess50K = Less50K['age'].mean()
StdDevLess50K = Less50K['age'].std()
MeanMore50K = More50K['age'].mean()
StdDevMore50K = More50K['age'].std()
print("Mean of age for those who earn more than 50K: ", MeanMore50K)
print("Standard Deviation of age for those who earn more than 50K: ", StdDevMore50K)
print("Mean of age for those who earn less than 50K: ", MeanLess50K)
print("Standard Deviation of age for those who earn less than 50K: ", StdDevLess50K)
```

**6. Is it true that people who earn more than 50K have at least high school education? (*education â€“ Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* or *Doctorate* feature)**

```python
# You code here
print("# of people that earn more than 50K: ", More50K["age"].count())
More50KHigh = More50K[(More50K['education'] == 'HS-grad') | (More50K['education'] == 'Bachelors') |
                     (More50K['education'] == 'Prof-school') | (More50K['education'] == 'Assoc-acdm') | 
                     (More50K['education'] == 'Assoc-voc') | (More50K['education'] == 'Masters') | 
                     (More50K['education'] == 'Doctorate')]
print("# of people that earn more than 50K & Have at least High School: ", More50KHigh["age"].count())
print("Therefore, people who earn more than 50K not necessarily have at least High School Education")
```

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
# You code here
SingleMen = data[(data['marital-status'] == 'Divorced') | (data['marital-status'] == 'Widowed') |
                (data['marital-status'] == 'Never-married') | (data['marital-status'] == 'Separated')]
SingleMen['age'].count()
MarriedMen = data[(data['marital-status'] == 'Married-AF-spouse') | (data['marital-status'] == 'Married-civ-spouse') |
                (data['marital-status'] == 'Married-spouse-absent')]
MarriedMen['age'].count()

SMenG50K = SingleMen[SingleMen['salary'] == '>50K']
MMenG50K = MarriedMen[MarriedMen['salary'] == '>50K']

print("Proportion of single men than earn more than 50K: ", 
      (SingleMen50K["age"].count()/SingleMen['age'].count())*100, "%")
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
# You code here
MaxHours = data['hours-per-week'].max()
People99Hours = data[data['hours-per-week'] == 99]
People99HoursG50K = People99Hours[People99Hours['salary'] == '>50K']

print("The maximum number of hours a person works per week is: ", MaxHours)
print("The number of people who work 99 hours per week is: ", People99Hours['hours-per-week'].count())
print("The percentage of people who work 99 hours per week and earn more than 50K is: ",
      People99HoursG50K['salary'].count()/People99Hours['salary'].count()*100, '%')
```
