```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from wordcloud import WordCloud 
import plotly.graph_objs as go
from plotly.offline import iplot
```

```python
df = pd.read_csv("pokemon.csv") #descargar el archivo en drive
df.head() #visualizamos nuestros datos
```

# Word Cloud

```python
 tipo = df['Type 1'].value_counts().index
```

```python
type_cloud = WordCloud(background_color="white", width=1050, height=700).generate(" ".join(tipo))
plt.imshow(type_cloud)
plt.axis("off")
plt.show();
```

# Box Plot (plotly)

```python
def make_box(data, name, color):
    return go.Box(y=data, name=name, marker={"color":color})

hp = make_box(df.HP, "HP", "#BA55D3")
attack = make_box(df.Attack, "Attack", "#5C871B")
defense = make_box(df.Defense, "Defense", "#742987")
speed = make_box(df.Speed, "Speed", "#A1D455")
iplot([hp, attack, defense, speed])
```

# Heatmap

```python
f,ax = plt.subplots(figsize=(10, 9))
sns.heatmap(df.corr(), annot=True, linewidths=.5, fmt=".1f", ax=ax, cmap="Purples");
```
